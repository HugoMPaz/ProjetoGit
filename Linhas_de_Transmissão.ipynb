# Equações das linhas de transmissão Hayt 10.2

---

<figure>
    <img src="https://img001.prntscr.com/file/img001/ZpAt-yVxTUmDSCcUKRx7eA.png">
</figure>

Analisando o circuito que representa uma linha de transmissão geral utilizando a lei de kirchhoff das malhas e corrente obtemos:
$$V(z,t) -R\Delta zI(z,t)-L\Delta z \frac{\partial I(z,t)}{\partial t}=V(z+\Delta z,t)$$

$$I(z,t)-G\Delta zV(z+\Delta z,t)-C\Delta z\frac{\partial V(z+\Delta z,t)}{\Delta t}=I(z+\Delta z,t)$$

Organizando as duas equações e aplicando o limite de $\Delta z \to 0$ obtemos equações diferenciais parciais acopladas no dominio do tempo

$$\frac{\partial V(z,t)}{\partial z}=-RI(z,t)-L\frac{\partial I(z,t)}{\partial t}$$

$$\frac{\partial I(z,t)}{\partial z}=-GV(z,t)-C\frac{\partial V(z,t)}{\partial t}$$

A forma fasorial é definida como:

$$V(z,t)=Re\{V_{s}(z)e^{j\omega t}\}$$
$$I(z,t)=Re\{I_{s}(z)e^{j\omega t}\}$$

Derivando as componentes no dominio da frequencia 1x

$$\frac{dV_{s}(z)}{dz}=-[R+j\omega L]I_{s}(z)$$
$$\frac{dI_{s}(z)}{dz}=-[G+j\omega C]V_{s}(z)$$

Com a segunda derivada podemos relacionar as 2 expressões e encontrar as constantes de propagação

$$\frac{d^{2}V_{s}(z^)}{dz^2}=-[R+j\omega L]\frac{dI_{s}(z)}{dz}$$
$$\frac{d^{2}I_{s}(z)}{dz^2}=-[G+j\omega C]\frac{dV_{s}(z)}{dz}$$

Combinando os 2 pares de equações é possivel desacoplar as equações

$$\frac{d^{2}V_{s}(z)}{dz^2}=[R+j\omega L][G+j\omega C]V_{s}(z)=\gamma^{2}V_{s}(z)$$
$$\frac{d^{2}I_{s}(z)}{dz^2}=[G+j\omega C][R+j\omega L]I_{s}(z)=\gamma^{2}I_{s}(z)$$

A constante de propagação $\gamma$ é composta por uma $\alpha$ e $j\beta$

$$\gamma=\alpha+j\beta =\sqrt{(R+j\omega L)(G+j\omega C)}$$

Onde $\beta$ representa a constante de fase e $\alpha$ a constante de atenuação


$$\alpha = \frac{1}{\sqrt{2}}\sqrt{RG-\omega^{2}LC+[(R^2+\omega^2L^2)(G^2+\omega^2C^2)]^{\frac{1}{2}}}$$

$$\beta = \frac{1}{\sqrt{2}}\sqrt{-RG-\omega^{2}LC+[(R^2+\omega^2L^2)(G^2+\omega^2C^2)]^{\frac{1}{2}}}$$

obtendo o valor de $\beta$ é possivel determinar comprimento de onda $\lambda$ e velocidade de propagação $u$

$$\lambda=\frac{2\pi}{\beta}$$

e

$$u=\frac{\omega}{\beta}=f\lambda$$

 # Impedancia caracteristica **$Z_0$**  


---



A **impedancia caracteristica** $Z_{0}$ da linha é a razão entre a onda de tensao e a onda de corrente, que se propagam no sentido positivo, em qualquer ponto da linha






$$Z_{0}=\sqrt{\frac{R+jwL}{G+jwC}}=R_{0}+jX_{0}$$

A. Linhas sem perdas 
---
Uma **linha de transmissão** é dita **sem perdas** se os condutores da linha são perfeitos ($\sigma\approx\infty $) e o meio dieletrico que os separa é sem perdas ($\sigma \approx 0$)

Manipulando a equação geral das linhas de transmissão obtemos os seguintes resultados adotando os parametros sem perdas

$$ \frac{\partial^{2}V}{\partial{z^{2}}}=LC\frac{\partial^{2}V}{\partial{t^{2}}}$$

$$ \frac{\partial^{2}I}{\partial{z^{2}}}=LC\frac{\partial^{2}I}{\partial{t^{2}}}$$


Exemplo 11.1 Sadiku

---

Uma linha de transmissão sem perdas no ar tem impedancia caracteristica $Z_{0} $ $ [\Omega]$ e constante de fase B [rad/m] a uma frequencia X [MHz]. Calcule a indutancia por metro e a capacitancia por metro da linha. 

```python
# exemplo 11.1 apresenta o calculo de indutancia e capacitancia 
# conhecendo sua frequencia de operação, impedancia e constante de fase B
from ipywidgets import interact, interactive
# Importa a biblioteca numpy para vetores e matrizes
import numpy as np
# Importa a biblioteca de constantes fisicas
import scipy.constants as con
def Q1(X,B,R0):   
  X=X*1e6 # conversão de grandeza de entrada
  w = 2*np.pi*X # Frequencia angular
  C=B/(w*R0) # calculo da capacitancia LT sem percas
  L=R0**2*C # calculo da indutancia LT sem percas
  print("C: ",C,"F/m ","L: ",L,"H/m ")

w = interactive(Q1, X=(50.,150.), B=(1.,5.),R0=(40.,100.))
display(w)
```

B. Linha sem distorção
---
Uma **linha sem distorção** é uma linha na qual a constante de atenuação $\alpha$ é independente da frequencia, enquanto que a constante de fase $\beta$ é linearmente dependente da frequencia

Uma linha sem distorções tem Z $\Omega$, A mNp/m, uc, onde c é a velocidade da luz no vácuo, Encontre R,L,G,C e lambda a uma frequencia F MHz

```python
# exemplo 11.2 apresenta o calculo de Indutancia, Capacitancia, Condutancia, Resistencia 
# e comprimento de onda, conhecendo sua frequencia de operação F, constante de atenuacao A
#,sua u velocidade de propagação e impedancia caracteristica Z.
from ipywidgets import interact, interactive
# Importa a biblioteca numpy para vetores e matrizes
import numpy as np
# Importa a biblioteca de constantes fisicas
import scipy.constants as con
def Q2(Z,F,A,u):   
  F=F*1e6 # conversão de grandeza de entrada
  w = 2*np.pi*F # Frequencia angular
  u=(u*con.c)/10 # conversao de velocidade
  A=A*1e-3 # conversao de escala
  R= A*Z # calculo da resistencia
  L=Z/u # calculo indutancia
  G=A**2/R # calculo condutancia
  C= 1/(u*Z) # calculo capacitancia
  lamb=u/F # calculo comprimento de onda
  print("R: ",R,"Ohm ","L: ",L,"H/m ","G: ",G,"S/m ","\n C: ",C,"F/m ","lambda: ",lamb,"m ")

w = interactive(Q2, Z=(30.,90.), F=(50.,150.),A=(10.,30.),u=(3.,9.))
display(w)
```

# Impedancia de Entrada
A impedancia de entrada $Z_{ent}$ consiste numa manipulação algebrica dos parametros do circuito analisado considerando o comprimento $l$ da linha de transmissão, sua impedancia caracteristica $Z_0$ e uma carga conectada $Z_c$

$$Z_{ent}=Z_{0}\Big{[}\frac{Z_{c}+Z_{0}tgh(\gamma l)}{Z_{0}+Z{c}tgh(\gamma l)}\Big{]}$$

onde $\gamma$ é a constate de Propagação

A. Linha em curto ($Z_{c}=0$)
---
A equação da impedancia de entrada se torna:
$$ Z_{cc}=Z_{ent} \Big|_{Zc=0}=jZ_{0}tg\beta l
$$
considerando que é uma linha de transmissão sem perdas


O coeficiente de reflexão na carga é expresso por:
$$\Gamma_{c}=\frac{0-Z_0}{0+Z_0}=-1 $$


B. Linha em circuito Aberto ($Z_{c}=\infty)$
---
A equação da impedancia de entrada se torna:
$$ Z_{ca}=\lim_{Z_{ca}\to\infty} Z_{ent} = \frac{Z_{0}}{jtg \beta l} = -jZ_{0}cotg\beta l
$$
Com o coeficiente de reflexão da carga expresso como:
$$\lim_{Z_{ca}\to\infty}\Gamma_{C}=1$$


Exemplo 11.3 Certa linha de transmissao, que opera a $\omega = 10^{6}rad/s $, tem $\alpha = 8 dB/m$, $\beta = 1 rad/m$ e $Z_{0} = 60 +j40$ $\Omega$ e tem 2 m de comprimento. Se a linha esta conectada à uma fonte de 10 $\angle 0°$ V. $Z_{g} = 40 $ $\Omega$ e está terminada por uma carga $20 + j50$ $\Omega$, determine impedancia de entrada, corrente na entrada da linha e corrente na metade da linha 

```python
import numpy as np
# Importa a biblioteca de constantes fisicas
import scipy.constants as con
a = 8/8.686 # conversão de dB para Np pelo fator 8.686 dB
beta=1j # constate de fase
w= 1e6 # rad/s
Z0= 60+40j # Ohm
l= 2 # comprimento da linha em metros
Vg= 10 # tensao da fonte 
Zg=40 # Ohm
Zc= 20+50j # Ohm
gamma = a+beta # constante de propagação
Zent = Z0*(Zc+Z0*np.tanh(gamma*l))/(Z0+Zc*np.tanh(gamma*l))
Ient= Vg/(Zent+Zg)
Vent = Zent*Ient
Vplus0=0.5*(Vent+Z0*Ient)
Vminus0 = 0.5*(Vent-Z0*Ient)
Imeio= (Vplus0/Z0)*np.exp(-gamma*l/2)-(Vminus0/Z0)*np.exp(gamma*l/2)
print("Impedancia de entrada: ",Zent,"Ohm\nCorrente de entrada: ",abs(Ient)*1000,"<",180*(np.arctan(Ient.imag/Ient.real))/np.pi,"° mA\nCorrente na metade da linha: ",1000*abs(Imeio),"<",180*(np.arctan(Imeio.imag/Imeio.real))/np.pi,"° mA")
```

#C. Linha casada ($Z_{c}=Z_{0}$)

---

A linha casada representa que não há reflexão de onda na linha de transmissão pela condição de igualdade entra a impedancia caracteristica de linha e a impedancia da carga resultando uma coeficiente de reflexão igual a 0

$$\Gamma_{c}=0$$


Carta de Smith
---

A carta é em homenagem a Phillip Smith, um engenheiro da Bell Telephone Laboratories, quem inventou e refinou isso entre 1936 e 1939, enquanto trabalhava para compreender as linhas de transmissão e ondas estacionárias no que era considerado “altas frequências” de até 1 MHz (naquela época, chamadas de megaciclos por segundo). Este gráfico circular de aparência estranha se tornou a única ferramenta mais poderosa e útil para trabalhar e otimizar circuitos de alta frequência, com respeito às impedâncias de entrada e saída, mesmo em nossa época de computadores poderosos e ferramentas de projeto assistido por computador (CAD).

Entre as muitas utilidades, a carta de Smith fornece uma maneira eficiente de visualizar opções de projeto, quando tentar casar impedâncias entre estágios de fonte e carga, uma consideração muito importante em muitos circuitos, especialmente em projetos de RF.

A carta de Smith é construida baseando-se na equação do coeficiente de reflexão
$$\Gamma = \frac{Z_{c}-Z_{0}}{Z_{c}+Z_{0}}$$

Normalizando a equação por $Z_{0}$ ficamos com:
$$\Gamma=\Gamma_{r}+j\Gamma_{i}=\frac{z_{c}-1}{z_{c}+1}$$
Manipulando a equacao para isolar $\Gamma_{r}$ resultamos no circulo resistivo com
<center>

centro em ($\Gamma_{r}$,$\Gamma_{i}$)=$
(\frac{r}{1+r'},0)$

raio = $\frac{1}{1+r}$

Tecnicas de Casamento de impedancia
---

O objetivo de se buscar linhas de transmissão casadas com $\Gamma=0$ é reduzir as perdas por reflexão

Alguns tipos de rede de casamento


*   Transformador de quarto de onda (para cargas resistivas)
*   Casador de toco simples
*   Casador de toco duplo





Transformador de quarto de onda 
---
O transformador de quarto de onda consiste em encontrar uma impedancia caracteristica de linha de transmissão com base na sua impedancia caracteristica da linha e da carga com um comprimento equivalente a $\lambda /4$ de onda frequencia de operação, sendo aplicado a linhas de tranmissão sem perdas e com cargas puramente resistiva

<img src="https://img001.prntscr.com/file/img001/c_TEDfHbRkuRvfnjHFQTQg.png">

a equação resumida é apresentada como:
$$Z_{m}=\sqrt{Z_{0}Z_{L}}$$




Transformador de toco simples
---

Quando o casamento não é possivel com o transformador de quarto de onda, uma alternativa é o uso de pequenos trechos de linhas termiando em curto ou circuito aberto e posicionados em serie ou paralelo com a linha principal para efetuar o casamento


Transformador de toco duplo

```python
import plotly.graph_objects as go
from ipywidgets import interact, interactive
def smith(Z0,Zc,Xc):
    fig = go.Figure()
    Norm=(Zc+1j*Xc)/Z0
    fig.add_trace(go.Scattersmith(
        imag=[Norm.imag],
        real=[Norm.real],
        marker_symbol='x',
        marker_size=30,
        marker_color="green",
        subplot="smith1"
    ))
    Ynorm=1/Norm
    fig.add_trace(go.Scattersmith(
        imag=[Ynorm.imag],
        real=[Ynorm.real],
        marker_symbol='circle',
        marker_size=30,
        marker_color="pink",
        subplot="smith1"
    ))
    fig.update_layout(
    smith=dict(
        realaxis_gridcolor='red',
        imaginaryaxis_gridcolor='blue',
        domain=dict(x=[0,0.75])
    )
    )
    fig.update_smiths(bgcolor="lightgrey")

    fig.show()
w = interactive(smith, Z0=(0.1,10.), Zc=(0.1, 10.),Xc=(-100.,100.))
display(w)
```

# Analise Transiente
Coeficiente de reflexao na Carga é:
$$\frac{V_1^-}{V_1^+}=\Gamma_L=\frac{R_L-Z_0}{R_L+Z_0}$$

Coeficiente de reflexao na Fonte é 
$$\frac{V_1^-}{V_1^+}=\Gamma_g=\frac{R_g-Z_0}{R_g+Z_0}$$

Valor da Onda de tensao inicial é:
$$V_1^+=\frac{V_0Z_0}{R_g+Z_0}$$

O **coeficiente de reflexão da corrente**, em qualquer ponto da linha, é igual ao coeficiente de reflexao da tensao naquele ponto, com sinal negativo

```python
# Importa a biblioteca numpy para vetores e matrizes
import numpy as np
# Importa a biblioteca pyplot para plotagem
import matplotlib.pyplot as plt
def upulse(p,c): # Funcao Degrau unitario p = posição do degrau; c = comprimento do vetor
    z=np.zeros(p)
    o=np.ones(c-p)
    return np.concatenate((z,o),axis=None)

def z2g(Zl,Z0): # Coeficiente de Reflexao
    return (Zl-Z0)/(Zl+Z0)
```

```python
from ipywidgets import interact, interactive
def reflex(Z0,Zg,Zl):

      
#    Z0=50 # Impedancia da Linha
#    Zg=2*Z0 # Impedancia da Fonte
#    Zl=0.5*Z0 # Impedancia da Carga
    gG= z2g(Zg,Z0) # Coeficiente de reflexao da Fonte
    gL= z2g(Zl,Z0) # Coeficiente de reflexão da Carga
    u=2e8 # Velocidade de propagação
    L=200 # Comprimento da Linha
    t=L/u # Tempo que demora até o sinal chegar na carga percorrendo a Linha
    T=6*t # Periodo de Analise
    l=500
    S=np.linspace(0,T,l) # Componente X em us
    e=S%t
    Vr=0
    cont=1
    V0=27
    Vg=V0*(Z0/(Z0+Zg))*np.ones(l)
    Vl=np.zeros(l)
    p=np.zeros(6)
    for i in range(len(e)):
        if i>0:
      
                
            if e[i-1]>e[i]:
      
                if (cont%2)==0:
            
                    Vg=Vg+(1+gG)*Vr*upulse(i,l)
                    Vr=Vr*gG
                    
                if (cont%2)==1:
                    
                    if cont==1:
                        
                        Vl=(1+gL)*Vg*upulse(i,l)
                        Vr=gL*V0*(Z0/(Z0+Zg))
                        
                    else:
                        
                        Vl=Vl+(1+gL)*Vr*upulse(i,l)
                        Vr=Vr*gL

                p[cont-1]=Vr
                cont+=1
                
    plt.plot(S,Vg,label='Vg')
    plt.plot(S,Vl,label='Vl')
    plt.xlabel('tempo (s)')
    plt.ylabel('tensao (V)')
    plt.legend(loc="upper right")
    plt.title('Analise de Transiente')
w = interactive(reflex, Z0=(0.,100.), Zg=(0, 200.),Zl=(0.,50))
display(w)
```
